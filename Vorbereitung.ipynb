# Vorbereitung

---

Folgende Vorbereitungsschritte müssen gemacht werden, damit Ihr mit den Aufgaben anfangen könnt:
1.  GitHub Account erstellen
2.  Repository erstellen
3.  Git installieren
4.  Repository klonen
5.  Git Anmeldedaten einrichten

---

#### **1. GitHub Account erstellen**

[Registrierungslink](https://github.com/signup)

---

#### **2. Repository erstellen**

1. Auf [Github](https://github.com/) gehen
2. Oben links auf den *grünen* Button `New` klicken:
   
   ![New repository](images/new_repo.png)
  
1. Auf der nächsten Seite Folgendes machen und schließlich auf `Create repository` klicken:
    * `Repository Name` ausfüllen (Namenskonvention: `Gruppe<DeineGruppenNr>-<DeineKlasse>-Projekt`)
    * `Public` auswählen
    * Hacken bei `Add a README file` setzen

   ![Repo ausfüllen](images/repo_info.png)
  
4. Ihr landet auf Euere Repository Seite. Sie müsste ungefähr so aussehen:
   
   ![Erstellte Repo](images/created_repo.png)

Jetzt muss die Person, die die Repository erstellt hatte, den anderen Gruppenmitglied einladen. Dafür muss Folgendes gemacht werden:

1. In Euerem Repository klickt Ihr auf `Settings`, danach auf `Collaborators` auf der linken Seite und schließlich auf der *grünen* Button `Add people`

   ![Collaborator hinzufügen](images/collaborator_add.png)

2. Da trägst Du den GitHub Benutzernamen von Deinem Partner ein
3. Dein Partner wird dann im GitHub oben rechts auf die Glocke klicken müssen
4. Er landet dann auf folgende Seite und muss schließlich auf die Benachrichtigung klicken

   ![Collaborator Einladun](images/collaborator_invite.png)

5. Danach öffnet sich die Seite auf der die Einladung akzeptiert werden soll
   
   ![Collaborator accepted](images/accept_inv.png)

---

#### **3. Git installieren**

Installationsdatei liegt im Tausch.<br>
Im Installer einfach alles mit Standardeinstellungen installieren.<br>
VS Code nach der Installation neustarten.

---

#### **4. Repository klonen**

* Im Tausch liegt der ***Repositories*** Jupyter Notebook. Auf den Link gehen, der zu Deiner Gruppe gehört.<br>
  Auf den *grünen* Button `<> Code` klicken, im `Local` Reiter `HTTPS` auswählen und den angezeigten Pfad kopieren:

  ![Klonpfad kopieren](images/clone.png)
* Auf der linken Seite im VS Code auf `Source Control` Icon klicken, und schließlich auf den Button `Clone Repository`:

  ![Repository klonen](images/clone_repo.png)
* Mittig ganz oben im VS Code wird ein Eingabefenster erscheinen. Dort müsst Ihr den davor kopierten Link einfügen:
  
  ![Klonpfad einfügen](images/clone_paste_link.png)
* Einen Ordner auswählen in welchem die Repository geklont bzw. heruntergeladen wird (**HINWEIS**: Auf Desktop klonen, da es Probleme mit dem Netzlaufwerk gibt!)
* Zum Schluss erneut auf `Source Control` Icon klicken. Wenn alles korrekt gemacht wurde, müsste es so aussehen:

  ![Repo initialisiert](images/initialised.png)
    * Falls es nicht so aussieht, dann auf den Button `Initialize Repository` klicken. Dann müsste es korrekt aussehen

---

#### **5. Git Anmeldedaten einrichten**

1. Im VS Code ein neues Terminalfenster öffnen:
   
   ![Terminalfenster öffnen](images/terminal.png)
2. Folgende Befehle eingeben und entsprechend anpassen:
   ```
   git config user.email “Deine GitHub E-Mail”
   git config user.name “Dein GitHub Benutzername”
   ```
3. Folgenden Befehl eingeben, um zu überprüfen, ob alles korrekt eingestellt wurde:
   
   `git config –-list`

   Wenn alles korrekt eingerichtet wurde, müsste die Ausgabe so aussehen:

   ![Git Ergebnis](images/git_ergebnis.png)

---
